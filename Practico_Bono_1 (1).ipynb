#### MENOR DE DOS PARES: Escriba una función que devuelva el menor de dos números dados *si* ambos números son pares, pero devuelva el mayor si uno o ambos números son impares
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
def lesser_of_two_evens(a,b):
  if a % 2 == 0 and b % 2 == 0: ## Se saca el residuo para saber si los números son pares, si es =0, entonces son pares
    if a < b:
      return a                  ## Si a < b, dado que ambos son pares, retorne a
    else:
      return b
  elif a > b:                   ## Si no son pares ambos, y a es mayor que b, retorne a
    return a
  else:
    return b                    ## Por el contrario, b es mayor que a, retorne b
  pass
    #Colocar aqui la respuesta
  print(lesser_of_two_evens(2,4))
```

```python
# Revisión
lesser_of_two_evens(6,4)
```

```python
# Revisión
lesser_of_two_evens(2,5)
```

#### GALLETAS DE ANIMALES: Escribir una función que toma una cadena de dos palabras y devuelve True si ambas palabras comienzan con la misma letra
    animal_crackers('Ciudad Calor') --> True
    animal_crackers('Lugar Frio') --> False

```python
def animal_crackers(text):
  palabras = text.split()              ## Separa el texto por los espacios y los almacena en un vector de palabras, donde la primera posición hace referencia a las palabras y la segunda a las letras
  if palabras[0][0] == palabras[1][0]:
    return True
  else:
    return False
  pass
```

```python
# Revisión
animal_crackers('Ciudad Calor')
```

```python
# Revisión
animal_crackers('Lugar Frio')
```

#### MAYÚSCULAS: Escribir una función que haga mayúscula la primera y cuarta letra de una cadena de texto
     
    old_macdonald('macdonald') --> MacDonald
    
Note: `'macdonald'.capitalize()` returns `'Macdonald'`

```python
def old_macdonald(name):
  if len(name) > 3:
    return name[0].upper() + name[1:3] + name[3].upper() + name[4:] ## Toma desde la posición 0 hasta la 3, volviendo mayúscula la primera letra, y lo concatena con la palabra desde la posición 3 hasta el final, volviendo mayúsucula la primera letra del arreglo, es decir la de la posición 3
  else:
    name = name[0].upper() # Si la palabra no es mayor a 3 letras, entonces solo devuelve la palabra con la primera en mayúscula
    print("El nombre es muy corto, por lo que solo se le transformará la primera letra en mayúscula")
    return name
    pass
```

```python
# Check
old_macdonald('macdonald')
```

#### REVERSA: Dada una oración, retornar la misma oración pero en reversa

    master_yoda('I am home') --> 'home am I'
    master_yoda('We are ready') --> 'ready are We'
    
Note: Revisar método .join()

    >>> "--".join(['a','b','c'])
    >>> 'a--b--c'

Resultado de .join():

    >>> " ".join(['Hello','world'])
    >>> "Hello world"

```python
def master_yoda(text):
  palabras = text.split()     ## Separa el texto en un vector de palabras
  palabras.reverse()          ## Intercambia las posiciones del vector de tal modo que quede en reversa
  return " ".join(palabras)   ## Une todo el vector, poniendo espacios entre cada posición
  pass
```

#### PROBLEMA 33:

Dada una lista de enteros, retornar True si el arreglo tinene un  3  seguido por un  3 .

    has_33([1, 3, 3]) → True
    has_33([1, 3, 1, 3]) → False
    has_33([3, 1, 3]) → False

```python
def has_33(nums):
  for i in range(0, len(nums)-1):           ## creación del iterador desde 0 hasta la longitud del arreglo menos uno (pues se inicia desde la posición 0)
    if nums[i] == 3 and nums[i+1] == 3:     ## Si la posición actual y la siguiernte son iguales a 3...
     return True                            ## Retorne True
  return False
  pass
```

```python
# Revisión
has_33([1, 3, 3])
```

```python
# Revisión
has_33([1, 3, 1, 3])
```

```python
# Revisión
has_33([3, 1, 3])
```

#### REPLICADOR: Dada una cadena, devuelve una cadena donde por cada carácter en el original hay tres caracteres REPETIDOS
    paper_doll('Hello') --> 'HHHeeellllllooo'
    paper_doll('Mississippi') --> 'MMMiiissssssiiippppppiii'

```python
def paper_doll(text):
  salida = ''         ## Creación de variable resultado que se devolverá
  for l in text:      ## por cada letra dentro del texto...
    salida += l*3     ## Repítalo 3 veces, y añadalo a la salida.
  return salida
  pass
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

#### BLACKJACK: Dados tres enteros entre 1 y 11, si su suma es menor o igual a 21, devuelve su suma. Si su suma excede 21 *y* hay un once, reduce la suma total en 10. Finalmente, si la suma (incluso después del ajuste) excede 21, devuelve 'BUST'
    blackjack(5,6,7) --> 18
    blackjack(9,9,9) --> 'BUST'
    blackjack(9,9,11) --> 19

```python
def blackjack(a,b,c):
  suma = a+b+c                               ## Se saca el resultado de la suma
  if suma <= 21:                             ## Si la suma es menor o igual a 21...
    return suma                              ## Devuelva la suma
  else:
    if a == 11 or b == 11 or c == 11:        ## Sino, se revisa si alguno de las 3 letras es igual a 11
      suma = suma - 10                       ## Se le resta 10 haciendo referencia a la elasticidad de valores cuando uno saca A en el blackjack
    else:
      suma = 'BUST'                          ## Si no hay ningun valor 11 y la suma es mayor a 21, retorne 'BUST'
  return suma
  pass
```

```python
# Revisión
blackjack(5,6,7)
```

```python
# Revisión
blackjack(9,9,9)
```

```python
# Revisión
blackjack(9,9,11)
```
